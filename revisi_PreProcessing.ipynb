# Import Lib

```python
import pandas as pd
import numpy as np
import nltk
import string
import re
```

```python
df = pd.read_csv('HasilScraping.csv')
df.drop(columns=df.columns[0], axis=1, inplace=True)
df
```

# Cleaning Data

```python
#Pembersihan Data (Tahap 1)

#remove mention
def remove_pattern(tweet,pattern):
    r = re.findall(pattern,tweet)
    for i in r:
        tweet = re.sub(i, '',tweet)
    return tweet

df['tweet_clean'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")

def clean_data(tweet):    
    #remove hastag, link dan tandabaca ^dan kata yg menyertainya 
    tweet = ' '.join(re.sub('([#^][A-Za-z0-9]+)|(\w+:\/\/\S+)',' ',tweet).split())
    
    #remove non ASCII (ex:emoticon)
    tweet = tweet.encode('ascii','replace').decode('ascii') 
        
    #remove number 
    tweet = re.sub(r'\d+','', tweet)
        
    #remove multiple whitespace into single whitespace
    tweet = re.sub('\s+',' ',tweet)
    
    #remove whitespace leading & trailing
    tweet = tweet.strip()
    
    #replace word repetition ('memuaskannnn' menjadi 'memuaskann')
    tweet = re.sub(r'(.)\1+', r'\1\1',tweet)
    
    #remove single character
    tweet = re.sub(r"\b[a-zA-Z]\b","",tweet)
    
    #remove punctuation (tanda baca)
    remove = string.punctuation
    translator = str.maketrans(remove, ' '*len(remove))
    tweet = tweet.translate(translator)
    
    return tweet.replace('http://',' ').replace('https://',' ')

df['tweet_clean'] = df['tweet_clean'].apply(clean_data)

df.tail(5)
```

```python
# Cleaning Data (2)

def clean_data1(tweet):               
    #remove multiple whitespace into single whitespace
    tweet = re.sub('\s+',' ',tweet)
    
    #remove whitespace leading & trailing
    tweet = tweet.strip()
    
    return tweet

df['tweet_clean'] = df['tweet_clean'].apply(clean_data1)

df.tail(5)
```

# Case Folding

```python
def casefold_data(tweet):
    tweet=tweet.lower()
    return tweet
    
df['tweet_clean'] = df['tweet_clean'].apply(casefold_data)

df
```

```python
#drop duplicates
df.drop_duplicates(subset="tweet_clean", keep="first", inplace=True)
df
```

# Tokenizing

```python
from nltk.tokenize import word_tokenize

def word_tokenize_wrapper(text):
    return word_tokenize(text)

df['tweet_clean'] = df['tweet_clean'].apply(word_tokenize_wrapper)

df.head(5)
```

```python
df.drop(df.index[1], inplace=True)
df
```

# Normalization Word

```python
normalized_word = pd.read_csv("normalisasi.csv")
normalized_word_dict = {}

for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]]=row[1]
        
def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

df['tweet_normalize'] = df['tweet_clean'].apply(normalized_term)

df.head(5)
```

# Filtering (Stopword Removal)

```python
from nltk.corpus import stopwords 
list_stopwords = stopwords.words('indonesian')

#menambah stopword dari file csv
txt_stopword = pd.read_csv('stopword_id.csv', names= ["stopwords"], header = None)

# convert stopword string to list & append additional stopword
list_stopwords.extend(txt_stopword["stopwords"][0].split(' '))

# convert list to dictionary
list_stopwords = set(list_stopwords)

#nenghapus stopwords
def stopword_removal(words):
    return [word for word in words if word not in list_stopwords]

df['tweet_stopword'] = df['tweet_normalize'].apply(stopword_removal)

df.head(10)
```

```python
import stanza

# Inisialisasi Stanza dengan model bahasa Indonesia
stanza.download("id")
nlp = stanza.Pipeline("id")

# Fungsi untuk melakukan lemmatisasi pada daftar teks
def lemmatize_tweets(tweet_list):
    lemmatized_list = []
    
    for tweets in tweet_list:
        lemmatized_tweets = []
        for tweet in tweets:
            doc = nlp(tweet)
            lemmatized_words = [word.lemma for sent in doc.sentences for word in sent.words]
            lemmatized_tweet = ' '.join(lemmatized_words)
            lemmatized_tweets.append(lemmatized_tweet)
        
        lemmatized_list.append(lemmatized_tweets)
    
    return lemmatized_list

# Terapkan fungsi lemmatisasi pada kolom 'tweet'
df['lemmatized_tweets'] = df['tweet_stopword'].apply(lemmatize_tweets)

df
```

```python
import stanza
import pandas as pd

# Inisialisasi Stanza dengan model bahasa Indonesia
stanza.download("id")
nlp = stanza.Pipeline("id")

# Membuat DataFrame contoh dengan kolom 'tweet'
data = df['tweet_stopword']

df = pd.DataFrame(data)

# Fungsi untuk melakukan lemmatisasi pada teks tweet
def lemmatize_tweet(tweet_stopword):
    doc = nlp(tweet_stopword)
    
    # Lemmatisasi setiap kata dalam tweet
    lemmatized_words = [word.lemma for sent in doc.sentences for word in sent.words]
    
    # Gabungkan kata-kata yang telah dilematisasi kembali menjadi teks
    lemmatized_text = ' '.join(lemmatized_words)
    
    return lemmatized_text

# Melakukan lemmatisasi pada kolom 'tweet'
df['lemmatized_tweet'] = df['tweet_stopword'].apply(lemmatize_tweet)

# Tampilkan DataFrame dengan hasil lemmatisasi
print(df)
```

# Stemming Sastrawi (ID)

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict={}

for document in df['tweet_stopword']:
    for term in document:
        if term not in term_dict:
            term_dict[term]=' '
            
for term in term_dict:
    term_dict[term]=stemmed_wrapper(term)
    
#apply stemming term ke dataframe
def get_stemmed_term(document):
    return [term_dict[term]for term in document]
    
df['final_tweet'] = df['tweet_stopword'].swifter.apply(get_stemmed_term)
df.head(10)
```

# Post Stemming

```python
# combine token list to string documents

def remove_punct(text):
    text = ' '.join([char for char in text if char not in string.punctuation])
    return text

df['final_tweet'] = df['final_tweet'].apply(remove_punct)
df
```

```python
#remove duplicates

df.drop_duplicates(subset="final_tweet", keep="first", inplace=True)
df
```
